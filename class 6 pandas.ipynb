# apply own function on dataframe with multiple columns as input

```python
def obtained_marks(sub1,sub2,sub3,sub4,sub5):
    return sub1+sub2+sub3+sub4+sub5
obtained_marks(1,2,3,4,5)
```

```python
l1=[1,2,3,4,5]
obtained_marks(*l1)
```

```python
obtained_marks(l1[0],l1[1],l1[2],l1[3],l1[4])
```

```python
df.loc[:,'s1':'s5'].apply(obtained_marks)
```

```python
df.loc[:,'s1':'s5'].apply(lambda x:obtained_marks(*x),axis=1)
```

```python
def abc():
    return 'A','B','C'
abc()
```

```python
def grading_system(s1,s2,s3,s4,s5):
    obtained_marks=s1+s2+s3+s4+s5
    per=obtained_marks/500*100
    grade=""
    if per>=80:
        grade="A+"
    elif per>=70:
        grade="A"
    elif per>=60:
        grade="B"
    elif per>=50:
        grade="C"
    else:
        grade="Fail"
        
    return obtained_marks, per, grade
grading_system(88,87,98,90,91)
```

```python
df.loc[:,'s1':'s5'].apply(lambda x:grading_system(*x),axis=1)
```

```python
df["obtained marks"] = df.loc[:,"s1":"s5"].apply(lambda x:grading_system(*x), axis=1)
df
```

```python
df[["obtained marks","Per","Grade"]] = df.loc[:,"s1":"s5"].apply(lambda x:grading_system(*x), axis=1)
df
```

```python
df[["obtained marks","Per","Grade"]] = df.loc[:,"s1":"s5"].apply(lambda x:grading_system(*x), axis=1, result_type="expand")
df
```
